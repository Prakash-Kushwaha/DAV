# Matplotlib

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[2,3,4,5,6,8]

plt.plot(x,y,marker='h', ms=9,mfc='r', mec='b',lw=3, c='g', label='Time v/s Distance') # plots the graph

plt.title('My Graph', loc='left') # set graph title
plt.xlabel('Time') # set x-axis label
plt.ylabel('Distance') # set y-axis label
plt.grid(axis='x') # enables grid

plt.legend()
plt.show() # to show the graph
```

### **Note**

* `markersize` &rarr; ms
* `markeredgecolor` &rarr; mec
* `markerfacecolor` &rarr; mfc
* `linewidth` &rarr; lw
* `linestyle` &rarr; ls

# **Bar**

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y, color='hotpink', width=0.3)
plt.show()
```

## *Horizontal Bars*

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, color='r', height=0.3)
plt.show()
```

# **Histogram**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)
plt.hist(x, color='c')
plt.show()
```

```python
import matplotlib.pyplot as plt

data=[23,40,32,12]
mylabel=['Maths', 'Computer Science', 'Chemistry', 'Physics']
plt.pie(data, labels=mylabel, autopct='%.2f', startangle=0, explode=[0.3,0.1,0.1,0.1],
        shadow=True, colors=['r', 'g', 'b', 'c'])
plt.title('Pie Chart')
plt.legend(title = 'Marks')
plt.show()
```

# Pandas

```python
import pandas as pd
```

## *Pandas Series*

```python
data=['Raj', 'Aman', 'Satyam', 'Rahul']
column=pd.Series(data)
print(column)
```

```python
print(column[0])
```

```python
myvar=pd.Series(data, index=['alpha', 'beta','gamma', 'zeta'])
print(myvar)
```

```python
print(myvar['gamma'])
```

### *Key/Value Objects as Series*

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)
```

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index = ["day1", "day2"])
print(myvar)
```
